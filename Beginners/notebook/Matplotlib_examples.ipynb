```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(20)
y = [x_i + np.random.randn(1) for x_i in x]
a, b = np.polyfit(x, y, 1)
```

```python
plt.plot(x, y, 'o', np.arange(20), a*np.arange(20)+b, '-')
```
